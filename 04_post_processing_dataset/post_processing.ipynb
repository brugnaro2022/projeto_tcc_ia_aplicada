##### Transformando variáveis categóricas em numéricas
- Verificando quantidades dos valores das features categóricas, para analisarmos quais features tem a possibilidade de serem transformadas em numéricas

```python
def verify_features_by_type(df, type):
    if type == 'numeric':
        return df.select_dtypes(include=['int64', 'float64']).columns
    else:
        return df.select_dtypes(include=['object', 'category']).columns

def verify_feature_values_quantity(df, feature_list):
    for index, feature in enumerate(feature_list):
        quantity = len(df[feature].value_counts())
        print(f'{feature}: {quantity}')

def checking_possible_numerical_features(df, feature_list):
    for index, feature in enumerate(feature_list):
        quantity = len(df[feature].value_counts())
        if quantity < 20:
            print(f'{feature}: {quantity}')
            
def repĺace_categorical_to_numeric(dataset, feature, numeric_dic):
    return dataset[feature].replace(numeric_dic)
```

##### Verificando quais são as features categoricas

```python
categorical_features = verify_features_by_type(df, 'categorical')
```

```python
categorical_features
```

```python
verify_feature_values_quantity(df, categorical_features)
```

##### Features categóricas com soma de valores menores que 20 decidimos transformar em numéricas 

```python
checking_possible_numerical_features(df, categorical_features)
```
